## Часть 3. Обработка данных, фильтрация, агрегация

Представьте, что вам нужно получить только датафрейм с мангой, у которой больше 10 страниц. Для этого нужно пользоваться фильтрацией. В pandas это выглядит так: вы упоминаете название датафрейма, открываете квадратные скобки и прописываете там условие.

```python
df[df['number_of_pages'] > 10]
```

**Задание 1: выведите мангу, у которой больше 2000 лайков**

Когда мы фильтруем данные, можно использовать сразу несколько условий. Для этого оба должны быть прописаны в квадратных скобках и каждое логическое условие должно быть обрамлено круглыми. Ниже мы выводим мангу, у которой меньше 1000 лайков **и** меньше 10 страниц.

```python
df[(df['likes'] < 1000) & (df['number_of_pages'] < 10)]
```

Аналогично можно прописать условие на логическое "или". **Задание 2:** выведите мангу, у которой больше 50 000 просмотров или больше 1000 лайков

Данные, которые вы получаете, можно отсортировать. Это происходит с помощью метода `sort_values()`. Если вы хотите получить данные по-убывающей, надо прописать аргумент `ascending=False`.

```python
df.groupby('author_name')['number_of_pages'].mean().sort_values(ascending=False)
```

Другой способ записать все это.

```python
df.groupby('author_name').mean('number_of_pages').sort_values(by='number_of_pages', ascending=False)
```

**Как сортировать данные сразу по нескольким условиям?** Использовать аргумент `by=` внутри метода `.sort_values()`, передав туда список. В ячейке ниже датафрейм сперва сортируется по `views`, после этого - по `likes`, а далее выводится от больших значений к меньшим (`ascending=False`).

```python
df.sort_values(by=['views', 'likes'], ascending=False)
```

### Альтернативные способы группировки и фильтрации: `query` и `agg`

Есть и альтернативные варианты группировкии и фильтрации, но они мне нравятся меньше. 

**Фильтрация**. За фильтрацию отвечает метод query. Тогда внутри нужно прописывать условия в одних кавычках. Ниже мы выводим мангу, у которой больше 40 тысяч просмотров **и** больше 10 страниц.

```python
df.query('views > 40000 & number_of_pages > 10')
```

Можно работать и с колонками, которые содержат строки. Если мы хотим проверить колонку на соответствие/не соответствие определенному значению, это значение нужно прописать в других по типу кавычках (если используете одинарные - то условия в двойных), иначе возникнет ошибка.

```python
df.query('series_name != "хентай Оригинальные работы"')
```

**Агрегация данных**. Считать определенные статистики в pandas можно и с помощью метода `.agg()` (или `.aggregate()`, это то же самое). Его синтаксис работает так – внутрь в словаре мы передаем пары ключ-значение. где **ключ - название колонки**, а **значение - статистики, которые мы хотим рассчитать**. Если хотим рассчитать несколько статистик, это нужно указывать через список (в качестве значения подаем список с нужными нам статистиками.

```python
df.groupby('author_name').agg({'likes':'mean'})
```

```python
df.groupby('author_name')['likes'].agg({'mean'}) # pandas работает по-разному, в том числе и так
```

В примере ниже мы рассчитываем разные статистики для разных авторов: для переменной `likes` - среднее и медиану; для переменной `views` - минимальной и максимальное значение.

```python
df.groupby('author_name').agg({'likes':['mean', 'median'],
                              'views':['min', 'max']})
```

Функции (например, `min`, `max`) можно прописывать как со скобками, так и без них.

```python
# вариант 1
df.groupby('author_name').agg({'number_of_pages':'min'})
```

```python
# вариант 2 - без скобок тоже работает
df.groupby('author_name').agg({'number_of_pages':sum})
```

## Т-тест для двух средних

> *А вот и момент истины! Действительно ли среднее количество просмотров для манги с эскплицитным насилием больше, чем для манги без эксплицитного насилия?*

Для реализации t-test Стьюдента для двух средних мы обращаемся к двум колонкам (могут быть разными по размеру), распаковываем получившиеся с помощью `stats.ttest_ind()` значения в `stat` - значение t-статистики и `p` - p-value.

**Для дальнейшей работы нам нужно создать два отдельных датафрейма.**

**Задание 1: создайте датафрейм `violence_True`, в который вы сохраните только те наблюдения из основного датафрейма `df`, для которого колонка `is_violence` принимает значение `True`**

**Задание 2: создайте датафрейм `violence_False`, в который вы сохраните только те наблюдения из основного датафрейма `df`, для которого колонка `is_violence` принимает значение `False`**

```python
violence_True = df[df['is_violence'] == True]
violence_False = df[df['is_violence'] == False]
```

`p-value` ниже свидетельствует о том, значимы или незначимы отличия между средним по двум выборкам. В социальных науках принято считать приемлимым для отвержения нулевой гипотезы (об отсутствии отличий между средними) уровень даже не `0.05`, а скорее `0.01` и ниже. 

Посмотрите на ячейку ниже. 

```python
from scipy import stats
stat, p = stats.ttest_ind(violence_True['views'], violence_False['views'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

**Задание 2.5: напишите логическую проверку для значения p. Округлите значение p-value до двух знаков после запятой. Если p меньше или равно 0.01, напечатайте строку `Нулевая гипотеза может быть отвергнута. Значение среднего значения этой статистики значимо отличается для манги с эксплицитным насилием и манги без эксплицитного насилия`**. 

**Если p больше 0.01, напечатайте строку `Нулевая гипотеза при p-value <значение p-value> на имеющихся данных пока не может быть отвергнута. Сейчас мы не можем говорить о значимых различиях по среднему значению этой статистики для манги с эксплицитным насилием и без эскплицитного насилия`**

```python
p = round(p, 2)
if p <= 0.01:
    print('Нулевая гипотеза может быть отвергнута. Значение среднего значения этой статистики значимо отличается для манги с эксплицитным насилием и манги без эксплицитного насилия')
else:
    print(f'Нулевая гипотеза при p-value {p} на имеющихся данных пока не может быть отвергнута. Сейчас мы не можем говорить о значимых различиях по среднему значению этой статистики для манги с эксплицитным насилием и без эскплицитного насилия')
```
